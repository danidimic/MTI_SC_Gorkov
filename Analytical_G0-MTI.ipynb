# Unperturbed Green's Function

We seek here the analytical solution for the _unperturbed_ normal propagator $\mathcal{G}^{(0)}$ that must be a solution of the differential equation 
$$
\Bigl( \hbar \omega - h_{MTI}(-i\partial_z) \Bigr) \mathcal{G}^{(0)}_{MTI}(z,z';\omega) = \hbar \delta(z-z') \,,
$$
where $h_{MTI}$ is the three-dimensional Hamiltonian for an MTI slab with finite thickness $d$ along the $z$ direction and fully translational invariant on the $x,y$ plane.



## **Analytical Solution**

In order to find an anlytical solution for the Green's function we assume $k_x=k_y=0$.
With this approximation, it is possible to show that only half of the components of the Green's fucntion are non-vanishing, namely those with equal spin $\sigma=\sigma'$
$$
	\mathcal{G}_{MTI}^{(0)} = 
	\begin{bmatrix}
		g_{\uparrow +, \uparrow +} & 0 & g_{\uparrow +, \uparrow -} & 0 \\[5pt]
		%
		0 & g_{\downarrow +, \downarrow +} & 0 & g_{\downarrow +, \downarrow -} \\[5pt]
		%
		g_{\uparrow -, \uparrow +} & 0 & g_{\uparrow -, \uparrow -}  & 0  \\[5pt]
		%
		0  & g_{\downarrow -, \downarrow +} & 0 & g_{\downarrow -, \downarrow -}
	\end{bmatrix} \,,
$$
where $g_{\sigma -, \sigma +} \equiv g_{\sigma -, \sigma +}(z,z';\omega)$.
Using the same procedure that we used for the numerical solution in the most general case, it is possible to compute the components of the above matrix aa
$$
    g_{\sigma \lambda, \sigma \lambda'}(z,z';\omega) =
    \begin{cases}
        \alpha_1(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z'-d)\lambda_1 \bigr\rbrack 
        \sinh \bigl( \lambda_1 z \bigr)
        +
        \alpha_2(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z'-d)\lambda_2 \bigr\rbrack 
        \sinh \bigl( \lambda_2 z \bigr) 
        \qquad 
        \text{for} \; z < z' \,, \\[10pt]
        %%%
        \alpha_1(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z-d)\lambda_1 \bigr\rbrack 
        \sinh \bigl( \lambda_1 z' \bigr)
        +
        \alpha_2(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z-d)\lambda_2 \bigr\rbrack 
        \sinh \bigl( \lambda_2 z' \bigr) 
        \qquad 
        \text{for} \; z < z' \,.
    \end{cases}
$$

The coefficients in the above equation are defined as following:

- i. $\lambda=+$, $\lambda'=+$
  
  $$
    \DeclareMathOperator{\csch}{csch}
    \alpha_1(\sigma, +, +) 
    = 
    \alpha \gamma_2 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, +, +)
    =
    -\alpha \gamma_1 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- ii. $\lambda=-$, $\lambda'=+$
  
  $$
    \alpha_1(\sigma, -, +)
    =
    \alpha \gamma_1 \gamma_2 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, -, +)
    =
    -\alpha \gamma_1 \gamma_2 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- iii. $\lambda=+$, $\lambda'=-$
  
  $$
    \alpha_1(\sigma, +, -)
    =
    -\beta 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, +, -)
    =
    \beta 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- iv. $\lambda=-$, $\lambda'=-$
  
  $$
    \alpha_1(\sigma, -, -)
    =
    -\beta \gamma_1
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, -, -)
    =
    \beta \gamma_2
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,.
  $$

In the equation above we defined $\alpha=\frac{\hbar}{B_1-D_1}$ and $\beta=-\frac{\hbar}{B_1+D_1}$, while the $\lambda_i$ and $\gamma_i$ parameters are defined as
$$
    \lambda_1 = \frac{1}{\sqrt{2}} \sqrt{A-\sqrt{A^2-4B}} \,,
    \qquad
    \lambda_2 = \frac{1}{\sqrt{2}} \sqrt{A+\sqrt{A^2-4B}} \,,
    \qquad
    \gamma_i = \lambda_i \bigl( C \lambda_i^2 - D \bigr) \,.
$$
In the above equation, we have introduced the following quantities
$$
    A = a+c+bcd \,, 
    \quad
    B = ac \,, 
    \quad
    C = \frac{1}{bc} \,, 
    \quad 
    D = \frac{a+bd}{bc} \,,
$$
which are defined in terms of the original Hamiltonian parameters as
$$
    a = -\frac{E_0^+}{B_1-D_1} \,, 
    \quad
    b = \frac{i p A_1}{B_1-D_1} \,,
    \quad
    c = \frac{E_0^-}{B_1+D_1} \,, 
    \quad
    d = \frac{-i p A_1}{B_1+D_1} \,,
$$
where
$$
    E_0^\pm = \epsilon_0 \pm \mu_0 + p \Lambda - \hbar \omega \,,
    \qquad
    p = 
    \begin{cases}
        +1 \qquad \text{for} \quad \sigma=\uparrow \,, \\[5pt]
        %
        -1 \qquad \text{for} \quad \sigma=\downarrow \,.
    \end{cases}
$$

```python
import numpy as np
from GreenFunctions_MTI import GMTI_modifiedBC, GMTI_normalBC, GMTI_NeumannBC, eigenstates

# Hamiltonian parameters
params=dict(C0 = -0.0068, D1 = 1.3, D2 = 19.6, A1 = 2.2, A2 = 4.1, M0 = 0.28, B1 = 10, B2 = 56.6)


# hyperbolic cosecant function
def csch(x):

    return 1. / np.sinh(x)


# function which determines the coefficients lambda_i and gamma_i
def parameters(irow, icol, L, omega, C0 = -0.0068, D1 = 1.3, A1 = 2.2, M0 = 0.28, B1 = 10, hbar = 1.):

    # switch over non-zero components
    match (irow,icol):
            
        # spin-up components
        case (0,0) | (0,2) | (2,0) | (2,2): p = +1

        # spin-down components
        case (1,1) | (1,3) | (3,1) | (3,3): p = -1
                                   
        # trivial components
        case _: p = 2

    
    # E0 plus/minus
    E0p = C0 + M0 + p*L - hbar*omega; E0m = C0 - M0 + p*L - hbar*omega
            
    # parameters a,b,c,d
    a = -E0p/(B1-D1); b = 1j*p*A1/(B1-D1); c = E0m/(B1+D1); d = -1j*p*A1/(B1+D1)

    # parameters A,B,C,D
    A = a+c+b*d; B = a*c; C = 1./(b*c); D = (a+b*d)/(b*c); Delta = np.sqrt(A**2-4*B)

    # lambda parameters
    lambda1 = np.sqrt((A-Delta)/2); lambda2 = np.sqrt((A+Delta)/2)
    # gamma parameters
    gamma1 = lambda1*( C*(lambda1**2)-D ); gamma2 = lambda2*( C*(lambda2**2)-D ) 
   
    return lambda1, lambda2, gamma1, gamma2



# function which define the coefficients alpha(sigma,lambda,lambda')
def coefficients(irow, icol, lambda1, lambda2, gamma1, gamma2, alpha, beta):

    # switch over non-zero components
    match (irow,icol):
            
        # (lambda, lambda') = (+,+)
        case (0,0) | (1,1):
            # coefficient alpha_1
            a1 = alpha*gamma2*csch(lambda1*d)/(gamma1*lambda2-gamma2*lambda1)
            # coefficient alpha_2
            a2 = -alpha*gamma1*csch(lambda2*d)/(gamma1*lambda2-gamma2*lambda1)
        
        # (lambda, lambda') = (-,+)
        case (2,0) | (3,1): 
            # coefficient alpha_1
            a1 = alpha*gamma1*gamma2*csch(lambda1*d)/(gamma1*lambda2-gamma2*lambda1)
            # coefficient alpha_2
            a2 = -alpha*gamma1*gamma2*csch(lambda2*d)/(gamma1*lambda2-gamma2*lambda1)

        # (lambda, lambda') = (+,-)
        case (0,2) | (1,3):
            # coefficient alpha_1
            a1 = beta*csch(lambda1*d)/(gamma1*lambda1-gamma2*lambda2)
            # coefficient alpha_2
            a2 = -beta*csch(lambda2*d)/(gamma1*lambda1-gamma2*lambda2)
                    

        # (lambda, lambda') = (-,-)
        case (2,2) | (3,3):
            # coefficient alpha_1
            a1 = beta*gamma1*csch(lambda1*d)/(gamma1*lambda1-gamma2*lambda2)
            # coefficient alpha_2
            a2 = -beta*gamma2*csch(lambda2*d)/(gamma1*lambda1-gamma2*lambda2)
                    
        # trivial components
        case _:
    
            a1 = 0.; a2 = 0.
                
    return a1, a2


# function evaluating the Green's function analytically for kx=ky=0
def GFanalytical(d, z, Z, L, omega, C0 = -0.0068, D1 = 1.3, A1 = 2.2, M0 = 0.28, B1 = 10, hbar = 1.):

    # non-homogeneous terms
    alpha = hbar/(B1-D1); beta = -hbar/(B1+D1)
    
    # define empty Green function
    GF = np.zeros((4,4), dtype='complex')

    # loop over rows
    for irow in range(4):
        
        # loop over columns
        for icol in range(4):

            # parameters depending on Hamiltonian 
            l1, l2, g1, g2 = parameters(irow, icol, L, omega, C0=C0, D1=D1, A1=A1, M0=M0, B1=B1, hbar=hbar)
            
            # coefficients for the Green's function components
            a1, a2 = coefficients(irow, icol, l1, l2, g1, g2, alpha, beta)

            # switch over non-zero components
            match (irow,icol):
            
                # g11 and g22
                case (0,0) | (1,1): 
                    
                    gij = a1*np.sinh(l1*z)*np.sinh(l1*(Z-d)) + a2*np.sinh(l2*z)*np.sinh(l2*(Z-d)) if z <= Z else a1*np.sinh(l1*Z)*np.sinh(l1*(z-d)) + a2*np.sinh(l2*Z)*np.sinh(l2*(z-d))

                # g31 and g42
                case (2,0) | (3,1): 
                    
                    gij = a1*np.cosh(l1*z)*np.sinh(l1*(Z-d)) + a2*np.cosh(l2*z)*np.sinh(l2*(Z-d)) if z <= Z else a1*np.cosh(l1*(z-d))*np.sinh(l1*Z) + a2*np.cosh(l2*(z-d))*np.sinh(l2*Z)

                # g13 and g24
                case (0,2) | (1,3): 
                    
                    gij = a1*np.sinh(l1*z)*np.cosh(l1*(Z-d)) + a2*np.sinh(l2*z)*np.cosh(l2*(Z-d)) if z <= Z else a1*np.sinh(l1*(z-d))*np.cosh(l1*Z) + a2*np.sinh(l2*(z-d))*np.cosh(l2*Z)

                # g33 and g44
                case (2,2) | (3,3): 
                    
                    gij = a1*np.cosh(l1*z)*np.cosh(l1*(Z-d)) + a2*np.cosh(l2*z)*np.cosh(l2*(Z-d)) if z <= Z else a1*np.cosh(l1*Z)*np.cosh(l1*(z-d)) + a2*np.cosh(l2*Z)*np.cosh(l2*(z-d))

                # trivial components
                case _:
                    
                    gij = 0.

            # fill the matrix for the Green's function
            GF[irow][icol] = gij

    return GF


# z range
d = 100.; Z = np.linspace(0., d, 250)


# Green's function analytical 
gf_analytical = np.array([GFanalytical(d=d, z=z, Z=z, L=0.005, omega=0.0025) for z in Z])
# Green's function with Neumann BC (diff. eq.)
gf_normalBC = np.array([GMTI_normalBC(d=d, z=z, Z=z, kx=0., ky=0., L=0.005, w=0.0025) for z in Z])
# Green's function with Neumann BC (diff. eq.)
gf_NeumannBC = np.array([GMTI_NeumannBC(d=d, z=z, Z=z, kx=0., ky=0., L=0.005, w=0.0025) for z in Z])
# Green's function with mixed BC (diff. eq.)
gf_modifiedBC = np.array([GMTI_modifiedBC(d=d, z=z, Z=z, kx=0., ky=0., L=0.005, w=0.0025) for z in Z])
```
