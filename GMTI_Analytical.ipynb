# Unperturbed Green's Function

We seek here the analytical solution for the _unperturbed_ normal propagator $\mathcal{G}^{(0)}$ that must be a solution of the differential equation 
$$
\Bigl( \hbar \omega - h_{MTI}(-i\partial_z) \Bigr) \mathcal{G}^{(0)}_{MTI}(z,z';\omega) = \hbar \delta(z-z') \,,
$$
where $h_{MTI}$ is the three-dimensional Hamiltonian for an MTI slab with finite thickness $d$ along the $z$ direction and fully translational invariant on the $x,y$ plane.



## **Analytical Solution**

In order to find an anlytical solution for the Green's function we assume $k_x=k_y=0$.
With this approximation, it is possible to show that only half of the components of the Green's fucntion are non-vanishing, namely those with equal spin $\sigma=\sigma'$
$$
	\mathcal{G}_{MTI}^{(0)} = 
	\begin{bmatrix}
		g_{\uparrow +, \uparrow +} & 0 & g_{\uparrow +, \uparrow -} & 0 \\[5pt]
		%
		0 & g_{\downarrow +, \downarrow +} & 0 & g_{\downarrow +, \downarrow -} \\[5pt]
		%
		g_{\uparrow -, \uparrow +} & 0 & g_{\uparrow -, \uparrow -}  & 0  \\[5pt]
		%
		0  & g_{\downarrow -, \downarrow +} & 0 & g_{\downarrow -, \downarrow -}
	\end{bmatrix} \,,
$$
where $g_{\sigma -, \sigma +} \equiv g_{\sigma -, \sigma +}(z,z';\omega)$.
Using the same procedure that we used for the numerical solution in the most general case, it is possible to compute the components of the above matrix aa
$$
    g_{\sigma \lambda, \sigma \lambda'}(z,z';\omega) =
    \begin{cases}
        \alpha_1(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z'-d)\lambda_1 \bigr\rbrack 
        \sinh \bigl( \lambda_1 z \bigr)
        +
        \alpha_2(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z'-d)\lambda_2 \bigr\rbrack 
        \sinh \bigl( \lambda_2 z \bigr) 
        \qquad 
        \text{for} \; z < z' \,, \\[10pt]
        %%%
        \alpha_1(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z-d)\lambda_1 \bigr\rbrack 
        \sinh \bigl( \lambda_1 z' \bigr)
        +
        \alpha_2(\sigma, \lambda, \lambda') 
        \sinh \bigl\lbrack (z-d)\lambda_2 \bigr\rbrack 
        \sinh \bigl( \lambda_2 z' \bigr) 
        \qquad 
        \text{for} \; z < z' \,.
    \end{cases}
$$

The coefficients in the above equation are defined as following:

- i. $\lambda=+$, $\lambda'=+$
  
  $$
    \DeclareMathOperator{\csch}{csch}
    \alpha_1(\sigma, +, +) 
    = 
    \alpha \gamma_2 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, +, +)
    =
    -\alpha \gamma_1 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- ii. $\lambda=-$, $\lambda'=+$
  
  $$
    \alpha_1(\sigma, -, +)
    =
    \alpha \gamma_1 \gamma_2 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, -, +)
    =
    -\alpha \gamma_1 \gamma_2 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- iii. $\lambda=+$, $\lambda'=-$
  
  $$
    \alpha_1(\sigma, +, -)
    =
    -\beta 
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, +, -)
    =
    \beta 
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,;
  $$

- iv. $\lambda=-$, $\lambda'=-$
  
  $$
    \alpha_1(\sigma, -, -)
    =
    -\beta \gamma_1
    \frac{\csch \bigl( \lambda_1 d \bigr)}
    {\lambda_1 (\gamma_1-\gamma_2)} \,,
    \qquad
    \alpha_2(\sigma, -, -)
    =
    \beta \gamma_2
    \frac{\csch \bigl( \lambda_2 d \bigr)}
    {\lambda_2 (\gamma_1-\gamma_2)} \,.
  $$

In the equation above we defined $\alpha=\frac{\hbar}{B_1-D_1}$ and $\beta=-\frac{\hbar}{B_1+D_1}$, while the $\lambda_i$ and $\gamma_i$ parameters are defined as
$$
    \lambda_1 = \frac{1}{\sqrt{2}} \sqrt{A-\sqrt{A^2-4B}} \,,
    \qquad
    \lambda_2 = \frac{1}{\sqrt{2}} \sqrt{A+\sqrt{A^2-4B}} \,,
    \qquad
    \gamma_i = \lambda_i \bigl( C \lambda_i^2 - D \bigr) \,.
$$
In the above equation, we have introduced the following quantities
$$
    A = a+c+bcd \,, 
    \quad
    B = ac \,, 
    \quad
    C = \frac{1}{bc} \,, 
    \quad 
    D = \frac{a+bd}{bc} \,,
$$
which are defined in terms of the original Hamiltonian parameters as
$$
    a = -\frac{E_0^+}{B_1-D_1} \,, 
    \quad
    b = \frac{i p A_1}{B_1-D_1} \,,
    \quad
    c = \frac{E_0^-}{B_1+D_1} \,, 
    \quad
    d = \frac{-i p A_1}{B_1+D_1} \,,
$$
where
$$
    E_0^\pm = \epsilon_0 \pm \mu_0 + p \Lambda - \hbar \omega \,,
    \qquad
    p = 
    \begin{cases}
        +1 \qquad \text{for} \quad \sigma=\uparrow \,, \\[5pt]
        %
        -1 \qquad \text{for} \quad \sigma=\downarrow \,.
    \end{cases}
$$

```python
import numpy as np

from GreenFunctions_MTI import GFexact

# Hamiltonian parameters
params=dict(C = -0.0068, D1 = 1.3, D2 = 19.6, A1 = 2.2, A2 = 4.1, M = 0.28, B1 = 10, B2 = 56.6)


# hyperbolic cosecant function
def csch(x):

    return 1. / np.sinh(x)


# function for alpha1 and alpha2
def alpha_coefficients(irow, icol, lambda1, lambda2, gamma1, gamma2, alpha, beta, d):

    a1 = 0; a2 = 0

    # matrix element (1,1) 
    if irow == 0 and icol == 0:
        # alpha1
        a1 = alpha*gamma2*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = -alpha*gamma1*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )
    
    # matrix element (3,1) 
    elif irow == 2 and icol == 0:
        # alpha1
        a1 = alpha*gamma1*gamma2*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = -alpha*gamma1*gamma2*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (2,2) 
    elif irow == 1 and icol == 1:
        # alpha1
        a1 = alpha*gamma2*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = -alpha*gamma1*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (4,2) 
    elif irow == 3 and icol == 1:
        # alpha1
        a1 = alpha*gamma1*gamma2*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = -alpha*gamma1*gamma2*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (1,3) 
    elif irow == 0 and icol == 2:
        # alpha1
        a1 = -beta*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = beta*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (3,3) 
    elif irow == 2 and icol == 2:
        # alpha1
        a1 = -beta*gamma1*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = beta*gamma2*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (2,4) 
    elif irow == 1 and icol == 3:
        # alpha1
        a1 = -beta*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = beta*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    # matrix element (4,4) 
    elif irow == 3 and icol == 3:
        # alpha1
        a1 = -beta*gamma1*csch(lambda1*d)/( lambda1*(gamma1-gamma2) )
        # alpha2
        a2 = beta*gamma2*csch(lambda2*d)/( lambda2*(gamma1-gamma2) )

    return a1,a2


# Function evaluating the Green's function analytically for kx=ky=0
def GFanalytical(d, z, Z, L, omega, C = -0.0068, D1 = 1.3, A1 = 2.2, M = 0.28, B1 = 10, hbar = 1.):

    # energies at k=0
    e0 = C; m0 = M

    GF = np.empty((4,4), dtype='complex')
    # loop over rows
    for irow in range(4):

        p = 0
        match irow:
            # first row, spin-up
            case 0: p=1
            # second row, spin-down
            case 1: p=-1
            # third row, spin-up
            case 2: p=1
            # fourth row, spin-down
            case 3: p=-1
        
        # loop over columns
        for icol in range(4):
            
            # E0 plus/minus
            E0p = e0 + m0 + p*L - hbar*omega; E0m = e0 - m0 + p*L - hbar*omega
            
            # parameters a,b,c,d
            a0 = -E0p/(B1-D1); b0 = 1j*p*A1/(B1-D1); c0 = E0m/(B1+D1); d0 = -1j*p*A1/(B1+D1)

            # parameters A,B,C,D
            Ap = a0+c0+b0*d0; Bp = a0*c0; Cp = 1./(b0*c0); Dp = (a0+b0*d0)/(b0*c0)

            # Delta, alpha and beta paramters
            Delta = np.sqrt(Ap**2-4*Bp); alpha = hbar/(B1-D1); beta = -hbar/(B1+D1)
            # gamma parameters
            lambda1 = np.sqrt(Ap-Delta)/np.sqrt(2); lambda2 = np.sqrt(Ap+Delta)/np.sqrt(2)
            # lambda parameters
            gamma1 = lambda1*( Cp*(lambda1**2)-Dp ); gamma2 = lambda2*( Cp*(lambda2**2)-Dp ) 
            
            # alpha coefficients
            alpha1, alpha2 = alpha_coefficients(irow, icol, lambda1, lambda2, gamma1, gamma2, alpha, beta, d)

            if z < Z:
                gij = alpha1*np.sinh((Z-d)*lambda1)*np.sinh(lambda1*z) + alpha2*np.sinh((Z-d)*lambda2)*np.sin(lambda2*z)
            
            else:
                gij = alpha1*np.sinh((z-d)*lambda1)*np.sinh(lambda1*Z) + alpha2*np.sinh((z-d)*lambda2)*np.sinh(lambda2*Z)

            # populate GF matrix
            GF[irow][icol] = gij

    return GF


#thickness
d = 100.
# z range
Z = np.linspace(0, d, 250)
# Green's function
gfzZ = np.array([GFanalytical(d=d, z=z, Z=z, L=0.005, omega=0.0025) for z in Z])

# compute exact Green's function on the lattice
gfexact = np.array([GFexact(d=d, z=z, Z=z, kx=0., ky=0., L=0.005, w=0.0025) for z in Z])
```

#### <font color='green'> **Real Part** </font> 

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

# colors for plotting
colors = ['brown', 'red', 'orange', 'gold']
# labels for curves
labels = [["$g_{\\uparrow +, \\uparrow +}$", "$g_{\\uparrow +, \downarrow +}$", "$g_{\\uparrow +, \\uparrow -}$", "$g_{\\uparrow +, \downarrow -}$"],
          ["$g_{\downarrow +, \\uparrow +}$", "$g_{\downarrow +, \downarrow +}$", "$g_{\downarrow +, \\uparrow -}$", "$g_{\downarrow +, \downarrow -}$"],
          ["$g_{\\uparrow -, \\uparrow +}$", "$g_{\\uparrow -, \downarrow +}$", "$g_{\\uparrow -, \\uparrow -}$", "$g_{\\uparrow -, \downarrow -}$"],
          ["$g_{\downarrow -, \\uparrow +}$", "$g_{\downarrow -, \downarrow +}$", "$g_{\downarrow -, \\uparrow -}$", "$g_{\downarrow -, \downarrow -}$"]]


# subplots
fig, axs = plt.subplots(2,2, figsize=(20, 20))

# indices referred to GF matrix
for icol in range(4):
        
    # indices for plotting
    if icol==0: iPic = 0; jPic = 0
    if icol==1: iPic = 0; jPic = 1
    if icol==2: iPic = 1; jPic = 0
    if icol==3: iPic = 1; jPic = 1
    
    for irow in range(4):
        # plot the Green's function
        axs[iPic][jPic].plot(Z, [gfzZ[idx][irow][icol].real for idx in range(len(Z))], c=colors[irow], lw=3, label=labels[irow][icol])
        # plot correct one
        axs[iPic][jPic].plot(Z, [gfexact[idx][irow][icol].real for idx in range(len(Z))], c=colors[irow], lw=3, ls=':')
    
    axs[iPic][jPic].set_xlabel("$z=z'$", fontsize = 20)
    axs[iPic][jPic].set_ylabel("$g_{\sigma\lambda, \sigma'\lambda'}(z,z;\omega)$", fontsize = 20)
        
    axs[iPic][jPic].tick_params(axis='both', which='major', labelsize=14)        
    axs[iPic][jPic].legend(fontsize=15);
```
